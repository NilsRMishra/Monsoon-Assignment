```python
# split payment history into delayed day values

def payment_string_split(x):
	return [int(x[i:i+3]) for i in range(0, len(x), 3)]

acc_data['payment_hist_string'] = acc_data['payment_hist_string'].apply(payment_string_split)
acc_data_t['payment_hist_string'] = acc_data_t['payment_hist_string'].apply(payment_string_split)
acc_data.head()
```

```python
# use the delayed days values to fill empties in close_date

today_date = pd.Timestamp.today().date()

def close_date_filler(item):
	if(pd.isna(item['closed_date'])):
		delta = len(item['payment_hist_string'])
		current_date = pd.to_datetime(item['open_date'])
		overdue_flag = item['amount_overdue']!=0
		if(overdue_flag):
			item['closed_date'] = str(today_date)
		else:
			item['closed_date'] = str((current_date + pd.DateOffset(months=delta)).date())
	return item['closed_date']
	
```

<h3>Binary Encoding of Nominal Features</h3>

```python
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
# adding binary encoded values for credit and enquiry types with prefix

binarizer_array = dict()
nominal_array = {'credit_type':'c', 'enquiry_type':'e'}
for col, pre in nominal_array.items():
	mlb = MultiLabelBinarizer()

	# encoding training data
	binarized_data = mlb.fit_transform(combined_data[col])
	binarized_df = pd.DataFrame(binarized_data, columns=mlb.classes_)
	binarized_df.columns = [f'{pre}_{col}' for col in binarized_df.columns]
	combined_data = pd.concat([combined_data, binarized_df], axis=1)

	# encoding testing data
	binarized_data = mlb.transform(combined_data_t[col])
	binarized_df = pd.DataFrame(binarized_data, columns=mlb.classes_)
	binarized_df.columns = [f'{pre}_{col}' for col in binarized_df.columns]
	combined_data_t = pd.concat([combined_data_t, binarized_df], axis=1)

	binarizer_array[col] = mlb
	print('done')
```

```python
def nested_min_max_avg(list_of_lists):
	flattened = [item for sublist in list_of_lists for item in sublist]
	if(len(flattened) == 0):
		return 0,0,0
	return min(flattened), max(flattened), sum(flattened) / len(flattened)

df = combined_data.iloc[0:5]
df['payment_hist_string'].apply(nested_min_max_avg)
```

```python
numerical_cols = ['loan_min', 'loan_max', \
       'loan_avg', 'due_min', 'due_max', 'due_avg', 'enq_min', 'enq_max', \
       'enq_avg', 'delay_min', 'delay_max', 'delay_avg']

binarized_cols = ['c_Another type of loan', 'c_Car loan', 'c_Cash loan (non-earmarked)', \
       'c_Consumer credit', 'c_Credit card', 'c_Interbank credit', \
       'c_Loan for business development', \
       'c_Loan for purchase of shares (margin lending)', \
       'c_Loan for the purchase of equipment', \
       'c_Loan for working capital replenishment', 'c_Microloan', \
       'c_Mobile operator loan', 'c_Mortgage', 'c_Real estate loan', \
       'c_Unknown type of loan', 'e_Another type of loan', 'e_Car loan', \
       'e_Cash loan (non-earmarked)', 'e_Cash loans', 'e_Consumer credit', \
       'e_Credit card', 'e_Interbank credit', \
       'e_Loan for business development', \
       'e_Loan for purchase of shares (margin lending)', \
       'e_Loan for the purchase of equipment', \
       'e_Loan for working capital replenishment', 'e_Microloan', \
       'e_Mobile operator loan', 'e_Mortgage', 'e_Real estate loan', \
       'e_Revolving loans', 'e_Unknown type of loan', 'is_loan_active']

date_cols = []
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score, f1_score, accuracy_score, classification_report, confusion_matrix
```

```python
# Normalize the numerical values

model_data = combined_data.copy()
test_data = combined_data_t.copy()
std = StandardScaler()
model_data[numerical_cols] = std.fit_transform(model_data[numerical_cols])
test_data[numerical_cols] = std.transform(test_data[numerical_cols])
model_data.head()
```

```python
model_data['TARGET'].value_counts()
```

```python
X = model_data[numerical_cols+binarized_cols]
y = model_data['TARGET']
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
train_df = pd.concat([X_train, y_train], axis=1)
train_majority = train_df[train_df['TARGET'] == 0]
train_minority = train_df[train_df['TARGET'] == 1]

# Resample Minor class
train_minority_upsampled = train_minority.sample(n=len(train_majority), replace=True, random_state=42)

# Resample both class
# train_minority_upsampled = train_minority.sample(n=len(train_majority)//4, replace=True, random_state=42)
# train_majority_downsampled = train_majority.sample(n=len(train_majority)//4, replace=True, random_state=42)

# Combine
# train_resampled = pd.concat([train_majority_downsampled, train_minority_upsampled])
train_resampled = pd.concat([train_majority, train_minority_upsampled])

# shuffle
train_resampled = train_resampled.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
X_train = train_resampled.drop(columns='TARGET')
y_train = train_resampled['TARGET']
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

y_pred = rf_model.predict(X_val)
```

```python
print(f"ROC_AUC Score for Random Forest Model : { roc_auc_score(y_pred, y_val)}")
```

```python
param_grid = {
    'max_depth': [3, 5],
    'learning_rate': [0.01, 0.1],
    'n_estimators': [100, 200],
    'subsample': [0.8, 1.0]
}
```

```python
xgbc = XGBClassifier(n_estimator=100, max_depth=5, min_child_weight=2, subsample=0.8, eval_metric='auc', booster='gbtree')
xgbc.fit(X_train, y_train)

y_preds = xgbc.predict(X_val)

print(f"ROC_AUC Score for Basic XGB Model : { roc_auc_score(y_preds, y_val)}")
```

```python
xgbc = XGBClassifier(eval_metric='auc', booster='gbtree')
grid_search = GridSearchCV(estimator=xgbc,
							param_grid=param_grid,
							scoring='roc_auc',
							cv=5,
							verbose=1,
							n_jobs=-1)
grid_search.fit(X_train, y_train)
```
